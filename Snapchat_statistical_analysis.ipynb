```python
import pandas as pd
import numpy as np

np.random.seed(42)  # for reproducibility

# Generate basic metrics
data = {
    'Impressions': np.random.randint(100, 10000, size=5000)
}

df = pd.DataFrame(data)
df['Clicks'] = df['Impressions'].apply(lambda x: np.random.randint(1, max(10, int(x * 0.1))))
df['Installs'] = df['Clicks'].apply(lambda x: np.random.randint(1, max(2, int(x * 0.5))))
df['Sales'] = df['Installs'].apply(lambda x: np.random.randint(0, max(1, int(x * 0.5))))
df['Revenue'] = df['Sales'].apply(lambda x: np.random.uniform(0.0, max(100.0, x * 500.0)))

# Derived metrics
df['CTR'] = df['Clicks'] / df['Impressions']
df['Install Rate'] = df['Installs'] / df['Clicks']
df['CPS'] = np.where(df['Sales'] > 0, df['Revenue'] / df['Sales'], 0)
df['ROAS'] = df['Revenue'] / (df['Impressions'] * 0.01)  # assuming $0.01 cost per impression

# Categorical variables
df['Gender'] = np.random.choice(['Male', 'Female', 'Other'], size=5000, p=[0.49, 0.49, 0.02])
df['Age_Group'] = np.random.choice(['18-24', '25-34', '35-44', '45-54', '55+'], size=5000)
df['Time_of_Day'] = np.random.choice(['Morning', 'Afternoon', 'Evening'], size=5000)

df.fillna(0, inplace=True)  # Handle any NaN

# Display the first few rows of the dataset
print(df.head())
```
